# Handling NAN Values




# For Floating Point Variables

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(missing_values = np.nan , strategy='mean')
```

Here we will fit all the null values with mean

```python
float_col = df.select_dtypes(include = 'float64').columns
```

```python
float_col
```

```python
si.fit(df[float_col])
```

```python
df[float_col] = si.transform(df[float_col])
```

Now we will devide our dependent and independent variables

```python
x = df.drop(columns = 'outcome')
y = df['outcome']
```

# For Dependent Variable

```python
Si = SimpleImputer(missing_values = np.nan , strategy='most_frequent')
```

```python
y = np.array(y).reshape(-1,1)
```

```python
Si.fit_transform(y)
```

```python
y= pd.DataFrame(y , columns = ['outcome'], dtype = 'int64')
```
